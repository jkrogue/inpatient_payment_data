```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colorbar import ColorbarBase
from mpl_toolkits.basemap import Basemap
import matplotlib as mpl

%matplotlib inline

pay_df = pd.read_csv("Inpatient_Prospective_Payment_System__IPPS__Provider_Summary_for_the_Top_100_Diagnosis-Related_Groups__DRG__-_FY2011.csv")

print(pay_df.head())
```

```python
# convert numeric columns to money
def mon_to_numeric(money_string):
    return float(money_string.split("$")[1])

cols_to_convert = [" Average Covered Charges ", " Average Total Payments ","Average Medicare Payments"]
converted_cols = ["avg_cov_chrg_num","avg_tot_pay_num","avg_medicare_pay_num"]

for i, each in enumerate(cols_to_convert):
    pay_df[converted_cols[i]] = pay_df[each].apply(mon_to_numeric)

print(pay_df[converted_cols].head())
```

```python
# Look for correlations

print(pay_df.corr())
```

**As expected, "average total payment" for an inpatient stay is related to "average covered charge" and "average medicare payment" but not related to anything else**

*Side note, can see strong negative correlation (r=-0.92) between longitude and zip code (i.e., as you move west zipcode increases)*

```python
# creating a dataframe groupby "DRG Definition"
drg_groups = pay_df.groupby("DRG Definition")

# Will sum up total discharges and average cost for each DRG
drg_df = drg_groups.agg({"avg_tot_pay_num":np.mean," Total Discharges ":np.sum})
drg_df.reset_index(inplace=True)
print(drg_df.head())
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(drg_df[[" Total Discharges ","avg_tot_pay_num"]],figsize=(20,15),diagonal="kde")
```

### Little relationship exists between total number of discharges and avg inpatient payment per DRG

```python
# Visualizing cost by DRG


#sort data by avg_tot_pay_num
drg_df.sort_values("avg_tot_pay_num",inplace=True)

print(drg_df.head())

fig, ax = plt.subplots(figsize=(20,10))
drg_df.plot.bar(x="DRG Definition",y="avg_tot_pay_num",ax=ax,legend=False)
ax.set_ylim(0,45000)
plt.title("Average total payment during inpatient stay per DRG",fontsize=20)
```

```python
#Plot most expensive DRGs

exp_drg_df = drg_df[-25:]

fig, ax = plt.subplots(figsize=(20,10))
exp_drg_df.plot.bar(x="DRG Definition",y="avg_tot_pay_num",legend=False,ax=ax)
ax.set_ylim(0,45000)
ax.set_title("Average total payment during inpatient stay for 25 most expensive DRGs",fontsize=20)
```

```python
#Plot cost from 25 common DRGs

common_drg_df = drg_df.sort_values(" Total Discharges ")
common_drg_df = common_drg_df[-25:]
common_drg_df.sort_values("avg_tot_pay_num",inplace=True)

fig, ax = plt.subplots(figsize=(20,10))
common_drg_df.plot.bar(x="DRG Definition",y="avg_tot_pay_num",ax=ax)
ax.legend().set_visible(False)
ax.set_ylim(0,45000)
ax.set_title("Average total payment during inpatient stay per DRG for 25 most common DRGs",fontsize=20)
```

# Notes on inpatient pmts per DRG

* Surgical DRGs are among most expensive
* Medical DRGs are among most common, but usually less expensive
* Joint replacement is the most expensive of the common DRGs

```python
# Visualizing most common DRGs

drg_df.sort_values(" Total Discharges ",inplace=True)
print(drg_df.head())

fig, ax = plt.subplots(figsize=(20,10))
drg_df.plot.bar(ax=ax,x="DRG Definition", y=" Total Discharges ", legend=False)
ax.set_title("Total number of inpatients discharged under each DRG",fontsize=20)
```

# ***Joint replacement is the most common DRG for an inpatient stay!*** 

```python
# Displaying DRGs with most varied avg payments

group = pay_df.groupby("DRG Definition")
std_df = group.agg({"avg_tot_pay_num":np.std})
std_df.reset_index(inplace=True)
std_df["stdev_tot_pay"] = std_df["avg_tot_pay_num"]
std_df.sort_values(by="stdev_tot_pay",inplace=True)

fig, ax = plt.subplots(figsize=(20,10))
std_df.plot.bar(x="DRG Definition", y="stdev_tot_pay",legend=False, ax = ax)
ax.set_title("Standard deviation of total payment per DRG between different providers",fontsize=20)
```

## Sepsis is the most variable charge, which makes sense.  Joint replacements with complications is high up as well

```python
# read in zip_codes
zipcodes = pd.read_csv("zip_codes.csv")
print(zipcodes.head())

# add leading zeros to zipcodes
#zipcodes["ZIP"] = zipcodes["ZIP"].apply(lambda x: str(x).zfill(5))
#print(zipcodes.head())
```

```python
# Merge zipcodes and pay_df to bring in latitudes and longitudes

pay_df["ZIP"] = pay_df["Provider Zip Code"]
print("pay_df.shape: %s\nzipcodes.shape: %s" % (pay_df.shape,zipcodes.shape))
pay_df = pay_df.merge(right=zipcodes,on="ZIP",how="inner")
print("Combined pay_df.shape: "  + str(pay_df.shape))
print(pay_df.head()[["ZIP","LAT","LNG"]])
```

```python
# average statistics by region
region_group = pay_df.groupby("Hospital Referral Region Description")
region_df = region_group.agg(np.mean)
region_df.reset_index(inplace=True)
print(region_df.head())
```

```python
print(region_df["avg_tot_pay_num"].sort_values(ascending=False))
```

```python
# visualizing average payment by region

plt.figure(figsize=(20,15))
ax1 = plt.subplot2grid((10,1),(0,0),rowspan=9)
ax2 = plt.subplot2grid((10,1),(9,0))
ax1.set_title("US regions color-coded by total avg inpatient payment",fontsize=20)
m = Basemap (projection="merc",llcrnrlon="-125",urcrnrlon="-67",llcrnrlat="25",urcrnrlat="50",)
m.scatter(ax=ax1, x=region_df["LNG"].tolist(),y=region_df["LAT"].tolist(),latlon=True,s=30,c=region_df["avg_tot_pay_num"],cmap="viridis")
m.drawmapboundary(ax=ax1)
m.drawcoastlines(ax=ax1,linewidth=1)
m.drawcountries(ax=ax1,linewidth=1)
m.drawstates(ax=ax1, linewidth=0.25,color="red")
cmap = mpl.cm.summer
min_tot = region_df["avg_tot_pay_num"].min()
max_tot = region_df["avg_tot_pay_num"].max()
norm = mpl.colors.Normalize(vmin=min_tot, vmax=max_tot)

cb1 = mpl.colorbar.ColorbarBase(ax2, cmap="viridis",
                                norm=norm,
                                orientation='horizontal')
cb1.set_label('Average total average payment')

#ColorbarBase(ax=ax2, cmap=mpl.cm.cool, norm=mp.colors.Normalize(vmin=0,vmax=20000), orientation="vertical")



```
